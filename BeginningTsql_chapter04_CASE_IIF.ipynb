###Exercise 4-6 (CASE & IIF)

01 (Simple CASE)

```sql
-- Select rows from a Table or View '[TableOrViewName]' in schema '[dbo]'
SELECT BusinessEntityID,
CASE BusinessEntityID % 2
WHEN 0 THEN 'Even'
ELSE 'Odd'
END AS [Odd Or Even]
FROM [HumanResources].[Employee];
```

02 (CASE)

```sql
-- Select rows from a Table or View '[TableOrViewName]' in schema '[dbo]'
SELECT OrderQty,


    CASE WHEN OrderQty BETWEEN 0 AND 9
THEN 'Under 10'
WHEN OrderQty BETWEEN 10 AND 19
THEN '10-19'
WHEN OrderQty BETWEEN 20 AND 29
THEN '20-29'
WHEN OrderQty BETWEEN 30 AND 39
THEN '30-39'
ELSE '40 and over'
END AS [Range]
FROM [Sales].[SalesOrderDetail];
```

06 (IIF)

```sql
-- Select rows from a Table or View '[TableOrViewName]' in schema '[dbo]'
USE WideWorldImporters
GO
SELECT DeliveryMethodID
, IIF(DeliveryMethodID IN (7,8,9,10), 'Freight',
'Courier/Other') AS [DeliveryMethod]
 FROM [Purchasing].[PurchaseOrders]

GO
```
